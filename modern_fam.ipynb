```python
# First version before finding out : errors
with open("season1.txt") as file_in:
    lines = []
    for line in file_in:
        if ':' in line: 
            cur_line = line.replace('\n','')
            lines.append(cur_line)
        # if it's a line spoken by the character, there is : quotation
        elif line == 'Modern Family\n':
            continue
        # get rid of Modern Family title
        elif '1x' in line:   
            continue
        # get rid of episode number
        else:
            prev_line = lines.pop()
            lines.append(prev_line + ' ' + line.replace('\n',''))
        # since the txt file 
```

```python
lines_values = [line.split(':') for line in lines]

df = pd.DataFrame(lines_values, columns=["Name", "LineA", "LineB", "LineC"])
df['Concatenated'] = df['LineA'].fillna('') + df['LineB'].fillna('') + df['LineC'].fillna('')
df = df.drop(columns = ['LineA','LineB','LineC'],axis = 1)
```
